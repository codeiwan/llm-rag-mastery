# Experiment 02: LangChain & ChatOpenAI Integration
- **Goal:** Verify the environment setup and basic connectivity of the LangChain-OpenAI integration.
- **Components:** `langchain-openai`, `python-dotenv`, `ChatOpenAI`.
- **Method:** Initialize the ChatOpenAI model and perform a simple synchronous invocation (`invoke`) to confirm response generation.

```python
# pip install langchain-openai==1.1.15 python-dotenv
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI()
```

```python
ai_message = llm.invoke("오늘 점심 메뉴를 딱 하나만 추천해주고, 그 메뉴를 골라준 이유를 설명해줘.")
```
